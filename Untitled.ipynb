```python
import boto3

s3 = boto3.resource('s3')
bucket_name = 'bucketchan'
prefix_amita = 'ML amita/'
prefix_reo = 'ML Reo/'

# Lista las imágenes en cada carpeta
bucket = s3.Bucket(bucket_name)
for obj in bucket.objects.filter(Prefix=prefix_amita):
    print(obj.key)
for obj in bucket.objects.filter(Prefix=prefix_reo):
    print(obj.key)
```

```python
from sagemaker.inputs import TrainingInput

train_data = TrainingInput(f's3://{bucket_name}/ML amita/', content_type='image/jpeg')
validation_data = TrainingInput(f's3://{bucket_name}/ML Reo/', content_type='image/jpeg')

```

```python
from sagemaker.inputs import TrainingInput

train_data = TrainingInput(f's3://bucketchan/ML amita/', content_type='image/jpg')
validation_data = TrainingInput(f's3://bucketchan/ML Reo/', content_type='image/jpg')
```

```python
from sagemaker.amazon.amazon_estimator import get_image_uri
import sagemaker

role = sagemaker.get_execution_role()

# Usa el contenedor de clasificación de imágenes
image_uri = get_image_uri(boto3.Session().region_name, 'image-classification')

estimator = sagemaker.estimator.Estimator(
    image_uri,
    role,
    instance_count=1,
    instance_type='ml.t3.medium',  # Usa 'ml.m5.large' si no tienes GPU
    volume_size=50,
    max_run=3600,
    input_mode='File',
    output_path=f's3://bucketchan/output/',
)

estimator.set_hyperparameters(
    num_layers=18,
    image_shape='3,224,224',
    num_classes=2,
    num_training_samples=1000,
    mini_batch_size=32,
    epochs=10,
    learning_rate=0.01,
)
```

```python
# Datos de entrenamiento y validación
data_channels = {
    'train': train_data,
    'validation': validation_data
}

# Inicia el entrenamiento
estimator.fit(data_channels)
```
